### #7.working with multiple stocks
+ We can download several stocks at the same time.
+ Create a list[] of stocks you want to work with.
+ Add the ticker symbols of the stock you want to get the information.
+ call the libary and variable add parameters if needed
  

```python
ticks = ["F","AAPL","AMZN","NFLX","GOOG"]
```

#### Notice: 
+ That first dimention is showing us the adj close(price) of all companies then closing price
+ Then the tickers we can simply chnage that by grouping(group_by) 

```python
yf.download(ticks, start="2020-01-01", end="2020-04-26")
```

```python
#used group_by to get more readable like format
yf.download(ticks, start="2020-01-01", end="2020-04-26",group_by="Ticker") 
```

### #8.Alteing the multi ticker columns 

```python
yf.download(ticks, start="2020-01-01", end="2020-04-26")["Adj Close"]
```

#### 8.1.Rounding method 
+ Make the data easy to visualize by rounding the numbers after the decimal point

```python
yf.download(ticks, start="2020-01-01", end="2020-04-26", rounding = True)["Adj Close"]
```

#### 8.2.Altering the list of tickers
+ By simply adding square brackets.
+ Type the ticker you intrested in.
+ It will return us with a series 

```python
yf.download(ticks, start="2020-01-01", end="2020-04-26", rounding = True)["Adj Close"] ["GOOG"]        
```
